```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
from scipy.stats.stats import pearsonr

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
# Create lists for data to be pulled
city_name_list = []
lat_list = []
lng_list =[]
max_temp_list = []
humidity_list = []
cloudiness_list = []
wind_speed_list = []
country_list = []
date_list = []
all_cities = []
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Save config information.
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"

# Build partial query URL
query_url = f"{url}appid={weather_api_key}&units={units}&q="
```

```python
# Loop through the list of cities and perform a request for data on each
# Print log of each city being processed with city number and name

index_tracker = 0

for index_tracker, city in enumerate(cities, start = 0):
      
    try:
        index_tracker = index_tracker + 1
        response = requests.get(query_url + city).json()
        city_name_list.append(response["name"])
        cloudiness_list.append(response["clouds"]["all"])
        country_list.append(response["sys"]["country"])
        date_list.append(response["dt"])
        humidity_list.append(response["main"]["humidity"])
        lat_list.append(response["coord"]["lat"])
        lng_list.append(response["coord"]["lon"])
        max_temp_list.append(response['main']['temp_max'])
        wind_speed_list.append(response["wind"]["speed"])

        print(f"Processing Record {index_tracker}: {city}") 
  
    except(KeyError, IndexError):
        print("City not found. Skipping...")

print("-----------------------------")
print("Data Retrieval Complete")
print("-----------------------------")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# create a dictionary for establishing dataframe
weather_dict = {
    "City":city_name_list,
    "Cloudiness":cloudiness_list,
    "Country":country_list,
    "Date":date_list,
    "Humidity":humidity_list,
    "Lat":lat_list,
    "Lng":lng_list,
    "Max Temp":max_temp_list,
    "Wind Speed":wind_speed_list 
}
```

```python
# Create a DataFrame 
weather_df = pd.DataFrame(weather_dict)

# Save Data to csv
weather_df.to_csv("../output_data/cities.csv")

# Display the DataFrame
weather_df.head()
```

## Linear Regression

```python
# OPTIONAL: Create a function to create Linear Regression plots

def linear_regression(x,y):
    print(f"The r-squared is : {round(pearsonr(x, y)[0],2)}")
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x, y)
    regress_values = x * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    plt.scatter(x, y)
    plt.plot(x,regress_values,"r-")
    return line_eq

# Define a fuction for annotating
def annotate(line_eq, a, b):
    plt.annotate(line_eq,(a,b),fontsize=15,color="red")
```

```python
# Create Northern and Southern Hemisphere DataFrames
northern_df = weather_df[(weather_df.Lat >= 0)]
northern_df.describe()
```

```python
# Create Northern and Southern Hemisphere DataFrames
southern_df = weather_df[(weather_df.Lat < 0)]
southern_df.describe()
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
equation = linear_regression(northern_df["Lat"], northern_df["Max Temp"])
annotate(equation, 0,0)
plt.title("Northern Hemisphere - City Latitude vs. Max Temperature Linear Regression (4 July 2020)")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (C)")
plt.grid (b=True,which="major",axis="both",linestyle="-",color="lightgrey")
plt.savefig("../output_data/Fig5.png")
plt.show()

# The linear regression shows the relationship between max temperature and latitude in the northern hemisphere.
# As expected, the further from equator (lat = 0), the cooler (lower) the max temperature.
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
equation = linear_regression(southern_df["Lat"], southern_df["Max Temp"])
annotate(equation, 0,0)
plt.title("Southern Hemisphere - City Latitude vs. Max Temperature Linear Regression (4 July 2020)")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (C)")
plt.grid (b=True,which="major",axis="both",linestyle="-",color="lightgrey")
plt.savefig("../output_data/Fig6.png")
plt.show()

# The linear regression shows the relationship between max temperature and latitude in the southern hemisphere.
# As expected, the closer to equator (lat = 0), the warmer the max temperature.
```
